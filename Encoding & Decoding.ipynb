```python
import pandas as pd
import pickle
```

```python
df=pd.read_csv("final_data.csv")
df
```

```python
df.info()
```

```python
# creating a list to hold name of the categorical columns
catg_col=df.select_dtypes(include="object").columns.to_list()
catg_col
```

```python
# one hot encoding
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder(sparse_output=False,handle_unknown="ignore")
encoder.fit(df[catg_col])
```

```python
encoder.categories_
```

```python
#list of new column
new_encoded_col= list(encoder.get_feature_names_out(catg_col))
new_encoded_col
```

```python
# transforming the data
df[new_encoded_col]=encoder.transform(df[catg_col])
df[new_encoded_col].head()
```

```python
df.head()
```

```python
# drop the unwanted columns
df.drop(columns=catg_col,inplace=True)
df
```

```python
with open('encoder.pkl','wb') as file:
    pickle.dump(encoder,file)
```
